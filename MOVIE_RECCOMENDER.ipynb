# **to get genre to the movies we want to convert this list of dictionary to list of genres as [Action,Adventure,Fantasy,Science Fiction]**

# index of dictionary is string we are converting it to list

```python
import ast
```

DEMO OF LITERAL EVALUATION

```python
l='[1,2,3,4,5]'
for i in ast.literal_eval(l):
  print(i)
```

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l;
convert(movies.iloc[0]['genres'])
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

# Similar preprocessing with keyword column

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

Similar preprocessing with Cast column
we will fetch name of first 3 actors as they only are the main actors we will ignore rest

```python

def convert2(obj):
  l=[]
  counter=0;
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l;
convert2(movies.iloc[0]['cast'])
```

# applying the function

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head()
```

for crew feature we will find only the name of the direction as it only effect the movie not the other behind the scene persons

```python
movies.iloc[0]['crew']
```

```python
def convert3(obj):
  l=[]

  for i in ast.literal_eval(obj):
      if i['job']=='Director':
       l.append(i['name'])
       break;

  return l;
convert3(movies.iloc[0]['crew'])
```

```python
movies['crew']=movies['crew'].apply(convert3)
```

```python
movies.head()
```

Converting string of overview to a list by spliting it from its spaces

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

we will remove spaces between name and surname of all the persons to avoid any confusion for the model for persons having same name or surname

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

# **now for further processing we will merge [overview	genres	keywords	cast_y	crew_y] columns into one to make a long paragraph and store it into a new paragraph named as tags**


```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
df=movies[['id','title','tags']]; # new processed dataframe
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x)) # converting list to string
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower()) # converting all letters to lower case
```

# **VECTORISATION**

**NOW WE WILL COVERT OUR TAGS COLUMN HAVING STRING TYPE DATA TO VECTOR BUT BEFORE THAT WE HAVE TO DELETE STOP WORDS FROM THE TAGS EG-> ARE, WE,WHERE,US,AS,FROM.**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
pd.DataFrame(vector)
```

```python
cv.get_feature_names_out
```

**from tags now we will delete same words such as action,actions,act these are differnet words but the meaning behind them is approximately same so we will convert all them into one**
# **THIS PROCESS IS KNOWN AS STEMMING**

```python
!pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(para):
  l=[];
  for i in para.split():
    l.append(ps.stem(i))
  return " ".join(l)
```

```python
df['tags']
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
df.head()
```

```python
vector=cv.fit_transform(df['tags']).toarray()
vector.shape
```

```python
print(cv.get_feature_names_out()[0:500])
```

```python
df.head()
```

**WE will calculate distance of each movie in the form of vector from all other movies in the form of vector
in this case we have very high dimensional data so we will not calculate euclidian distances instead we will calculate cosine distance

pictures having similarity -> value near to 1


dissimilar pictures -> value near to 0 **

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
```

```python
similarity=cs(vector)
```

```python
similarity
```

Sort the values of the similarity matrix to get most similar movies

```python
df.head(1)
```

FUNCTION WHICH TAKES THE MOVIE NAME AS THE INPUT AND RETURUNS THE 5 MOST MATCHING MOVIES FROM THE DATAST

```python
def recommend (movie):
  themovie=df[df['title']==movie].index[0]
  distances=similarity[themovie]
  movieslist=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] # enumerate to fetch the indices along with their similarity
  for i in movieslist:
    print(df.iloc[i[0]]['title'])
```

# **TESTING**

```python
recommend('Avatar') # reccomendations of movie Avatar
```
